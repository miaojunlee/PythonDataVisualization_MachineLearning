```python
import matplotlib.pyplot  as plt
%matplotlib inline
import numpy as np 
x= np.linspace(0,10)
y= x+(x*3)

fig= plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='Orange',linewidth=3,label='Legend is shown',alpha=0.5,linestyle="-",marker="x")

ax.legend(loc=0)  ### location code for best is 0. easiest and best location


```
