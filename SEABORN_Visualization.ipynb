```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
sns.distplot(tips['total_bill'],bins=40,kde=False)  ## distribution plot 
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',color='green',alpha=0.3,size=10,kind='hex')## joint plot 
```

```python
sns.pairplot(tips,size=5,hue='smoker')          ## hue is the best thing ever
```

```python
sns.kdeplot(tips['total_bill'])   ### not for columns holding string values. Only numeric values
```

```python
import numpy as np
sns.barplot(data=tips,x='sex',y='tip',alpha=0.3,estimator=np.std)
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',fliersize=8,hue='smoker')
```

```python
sns.violinplot(data=tips,x="day",y="total_bill")
```

```python
sns.countplot(x='sex',data=tips,alpha = 0.6) ## categorical plot 
```

```python
sns.stripplot(data=tips,x='day',y="total_bill",hue="sex",split=True)## split splits the two categories for data visulaization
```

```python
sns.factorplot(data=tips,x='day',y="total_bill",kind="violin")  ## kinds can be point, violin, bar and so on. It can take anyshape
```
