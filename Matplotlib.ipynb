```python
import matplotlib.pyplot  as plt
%matplotlib inline
import numpy as np 
x= np.linspace(0,5,11)
y= x**2
```

```python
plt.plot(x,y)  ####3 functional method
plt.xlabel('X numbers')
plt.ylabel('Y axis')
plt.title('My first matplot')
```

```python
### now we will use object oriented methods to plot 
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])  ## left, bottom, width, height)
axes1= fig.add_axes([0.2,0.4,0.3,0.45])
axes1.plot(x,y)
axes.plot(x,y)
axes.set_title('Obie Object Oriented plot')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')

####### Lets add another plot to this plot


```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```
