```python
import seaborn as sns
tips=sns.load_dataset("tips")
tips.head(5)
%matplotlib inline
```

```python
sns.lmplot(data=tips,x="total_bill",y="tip",size=10,hue="sex",markers=["X","v"])
```

```python
# instead of hue you can divide the grid in two. 
sns.lmplot(data=tips,x="total_bill",y="tip",size=8,hue="sex",col="day",aspect=0.7)
```
